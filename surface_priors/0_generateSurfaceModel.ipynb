```python
import os
from isofit.utils.surface_model import surface_model
from spectral.io import envi
from spectral.io.envi import read_envi_header
import matplotlib.pyplot as plt

%matplotlib widget
```

```python
surface_out = '/store/carroll/col/data/2018/surface_priors/'
emit_surface = '/store/carroll/emit_surface/'
os.listdir(emit_surface)
```

```python
# just to look at how they manually modified the spectra in preparation
fp = os.path.join(emit_surface, 'filtered_veg.hdr')
filtered_veg = envi.open(fp).open_memmap(interleave='bip')[:,0,:].copy()
wls = [int(x) for x in read_envi_header(fp)['wavelength']]

fig, ax = plt.subplots()
for i in range(5):
    ax.plot(wls, filtered_veg[i,:])
plt.show()
```
