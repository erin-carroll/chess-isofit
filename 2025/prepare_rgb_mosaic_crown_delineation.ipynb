```python
import os
from glob import glob
import h5py
import rasterio
from rasterio.merge import merge

import sys
sys.path.append('/store/carroll/repos/neon-isofit/')
from h5_to_envi import radianceH5toEnvi

os.chdir('/store/carroll/col/data/2025/')
```

```python
domains = ['2025_ALMO_1', '2025_CRBU_2', '2025_UPTA_1']

band_indexes = [60, 40, 30] 
```

```python
for domain in domains:
    print(domain)
    fps = [x for x in glob(f'raw/L3/reflectanceENVI/{domain}/*') if '.' not in x]
    fp_out = f'raw/L3/reflectanceENVI/CHESS25_{domain}_reflectance_rgb.tif'
    
    srcs = [rasterio.open(fp) for fp in fps]
    mosaic, out_transform = merge(srcs, indexes=band_indexes)
    meta = srcs[0].meta.copy()
    meta.update(
        driver="GTiff",
        height=mosaic.shape[1],
        width=mosaic.shape[2],
        count=len(band_indexes),
        transform=out_transform)

    with rasterio.open(fp_out, 'w', **meta) as dst:
        dst.write(mosaic)

    for src in srcs:
        src.close()
```

```python
# delete all of the gridded files once confirmed that rgb mosaics look good
for domain in domains:
    fol = f'/store/carroll/col/data/2025/raw/L3/reflectanceENVI/{domain}'
    os.rmdir(fol)
```
